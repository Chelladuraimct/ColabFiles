**Data Merging and Lookup Operations with Pandas**

---

```python
#Import Libraries
import pandas as pd
import numpy as np
```

---
**TASK 1: Data Preparation and Missing Value Handling**

---

**Create DataFrames**

```python
# Students Data
students_data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107],
    'name': ['Alice', 'Bob', None, 'David', 'Emma', 'Frank', 'Grace'],
    'email': ['alice@email.com', 'bob@email.com', 'charlie@email.com', None, 'emma@email.com', 'frank@email.com', 'grace@email.com'],
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Mumbai', None, 'Chennai', 'Delhi']
}

# Enrollments Data
enrollments_data = {
    'student_id': [101, 102, 103, 105, 108, 109],
    'course_name': ['Python', 'Data Science', 'Python', 'Machine Learning', 'AI', 'Python'],
    'enrollment_date': ['2024-01-15', '2024-01-20', '2024-02-01', '2024-02-10', '2024-02-15', '2024-03-01']
}

# Scores Data
scores_data = {
    'student_id': [101, 102, 104, 105, 106],
    'exam_score': [85, 92, 78, 88, 95]
}

# Convert to DataFrames
students_df = pd.DataFrame(students_data)
enrollments_df = pd.DataFrame(enrollments_data)
scores_df = pd.DataFrame(scores_data)

print("Original Students DataFrame:")
display(students_df)
```

---
**Null Value Analysis**

---

```python
print("Null Value Analysis:")

for column in students_df.columns:
    null_count = students_df[column].isnull().sum()
    percentage = (null_count / len(students_df)) * 100
    print(f"Column: {column}, Nulls: {null_count} ({percentage:.2f}%)")
```

---
**Data Cleaning**

---
**Fill Missing City Values**

```python
students_df['city'].fillna('Unknown', inplace=True)
```

**Drop Rows Where Name is Missing**

```python
students_df = students_df.dropna(subset=['name'])
```

**Display Cleaned DataFrame**

```python
print("Cleaned Students DataFrame:")
display(students_df)
print("Total Rows:", len(students_df))
```

---
**TASK 2: Multiple Join Operations**

---
**Inner Join**

```python
inner_join = pd.merge(students_df, enrollments_df, on='student_id', how='inner')

print("Inner Join Result:")
display(inner_join)

print("Total Students in Result:", len(inner_join))
```

Explanation

Students excluded: 104, 106, 107

Reason:
They do not exist in the enrollments table, so inner join removes them.

**Left Join**

```python
left_join = pd.merge(students_df, enrollments_df, on='student_id', how='left')

print("Left Join Result:")
display(left_join)

print("Total Rows:", len(left_join))

#Students With No Course
no_course = left_join[left_join['course_name'].isnull()]
print("\nStudents with NULL course_name:")
display(no_course[['student_id','name']])
```

Explanation:
Students 104, 106, 107 have no enrollments.

**Right Join**

```python
right_join = pd.merge(students_df, enrollments_df, on='student_id', how='right')

print("Right Join Result:")
display(right_join)

print("Total Rows:", len(right_join))

#Students Without Names
missing_names = right_join[right_join['name'].isnull()]
print("\nStudent IDs without names:")
display(missing_names[['student_id','course_name']])
```

Explanation:
Student IDs 108 and 109 exist in enrollments but not in students table.

**Full Outer Join**

```python
outer_join = pd.merge(students_df, enrollments_df, on='student_id', how='outer')

print("Outer Join Result:")
display(outer_join)

print("Total Rows:", len(outer_join))

#Rows With Missing Values
missing_rows = outer_join[
    (outer_join['name'].isnull()) |
    (outer_join['course_name'].isnull())
]

print("\nRows with Missing Name OR Course:")
display(missing_rows)

#Merge Indicator Analysis
outer_indicator = pd.merge(
    students_df,
    enrollments_df,
    on='student_id',
    how='outer',
    indicator=True
)

display(outer_indicator)

print("\nMerge Source Distribution:")
print(outer_indicator['_merge'].value_counts())
```

---
**TASK 3: Lookup Operation and Automation**

---
**Lookup Using Dictionary + map()**

```python
# Create dictionary
score_lookup = scores_df.set_index('student_id')['exam_score'].to_dict()

# Map scores
students_df['exam_score'] = students_df['student_id'].map(score_lookup)

print("Students with Exam Scores:")
display(students_df[['student_id','name','exam_score']])
```

Students without exam scores will show NaN.

**Same Operation Using Merge**

```python
merge_scores = pd.merge(
    students_df,
    scores_df,
    on='student_id',
    how='left'
)

display(merge_scores[['student_id','name','exam_score_y']])
```

**Simple Automation Function**

```python
def auto_merge(df1, df2, join_type, key_column):

    merged_df = pd.merge(
        df1,
        df2,
        on=key_column,
        how=join_type
    )

    result = {
        'result_df': merged_df,
        'row_count': len(merged_df),
        'join_type': join_type
    }

    return result
```
