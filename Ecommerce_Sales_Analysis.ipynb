**Ecommerce Sales Analysis using Pandas**

---
**Import Libraries and Create Synthetic Dataset**

**Explanation**

Since the company database cannot be accessed, we create a synthetic dataset using Python’s random module and store it in a Pandas DataFrame.

**The dataset contains:**

| Column         | Description               |
| -------------- | ------------------------- |
| transaction_id | Unique transaction number |
| region         | Sales region              |
| category       | Product category          |
| salesperson    | Person who made the sale  |
| sales_amount   | Value of sale             |
| customer_id    | Customer identifier       |


```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)

print(df.head(10))
print("\nDataset shape:", df.shape)
```

---
**Task 1: Basic Grouping and Single Aggregations**

---
**1.1 Total Sales by Region**

**Explanation**

We group the data by region and calculate the total sales amount using sum().


```python
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()

# Sort descending
region_sales = region_sales.sort_values(by='sales_amount', ascending=False)

print(region_sales)
```

**1.2 Count Transactions by Category**

**Explanation**

We count the number of transactions per product category.

```python
category_transactions = df.groupby('category')['transaction_id'].count().reset_index()

print(category_transactions)
```

**1.3 Average Sales per Salesperson**

**Explanation**

This calculates mean sales amount per salesperson.

```python
avg_salesperson = df.groupby('salesperson')['sales_amount'].mean().reset_index()

print(avg_salesperson)
```

---
**Task 2: Multi-Column Grouping and Multiple Aggregations**

---

**2.1 Total Sales by Region AND Category**

**Explanation**

Grouping by multiple columns allows deeper insights such as regional performance of each category.

```python
region_category_sales = df.groupby(['region','category'])['sales_amount'].sum().reset_index()

print(region_category_sales.head())
```

**2.2 Multiple Metrics per Salesperson**

**Explanation**

**Using agg() we calculate:**

*   Total Sales
*   Average Sales
*   Transaction Count






```python
salesperson_stats = df.groupby('salesperson')['sales_amount'] \
                      .agg(['sum','mean','count']) \
                      .reset_index()

# Sort by total sales
salesperson_stats = salesperson_stats.sort_values(by='sum', ascending=False)

print(salesperson_stats)
```

**2.3 Category with Maximum Revenue**

**Explanation**

We calculate total sales per category and use idxmax() to find the category with the highest revenue.

```python
category_sales = df.groupby('category')['sales_amount'].sum()

top_category = category_sales.idxmax()

print("Category with maximum revenue:", top_category)
```

---
**Task 3: Custom Aggregation and Complete Sales Report**

---
**3.1 and 3.2 Custom Aggregation Function**

**Explanation**

We create a custom function to calculate the range of sales values:

range = max - min

```python
def sales_range(x):
    return x.max() - x.min()

region_analysis = df.groupby('region')['sales_amount'] \
                    .agg(['sum','mean','min','max',sales_range]) \
                    .reset_index()

print(region_analysis)
```

**3.3 Final Summary Sales Report**

**Explanation**

We generate a region-wise sales report including:

*   Total sales
*   Average sales
*   Number of transactions

```python
final_report = df.groupby('region').agg({
    'sales_amount': ['sum','mean'],
    'customer_id': 'count'
}).reset_index()

print(final_report)
```

**3.4 Explanation: Multi-Level Columns in Aggregation**

When using dictionary-based aggregation, Pandas produces multi-level column names.

**Example:**

sales_amount

    sum

    mean

customer_id

    count

This creates a hierarchical column structure:

| Level 1      | Level 2 |
| ------------ | ------- |
| sales_amount | sum     |
| sales_amount | mean    |
| customer_id  | count   |

**Difference from Single Aggregation**

**Single aggregation:**

```python
df.groupby('region')['sales_amount'].sum()
```

**Multi aggregation:**

```python
df.groupby('region').agg({
'sales_amount':['sum','mean']
})
```
