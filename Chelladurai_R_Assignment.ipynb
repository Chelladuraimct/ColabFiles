Step 1: Create the Data Array

```python
import numpy as np

data = np.array([[22.5, 19.0, 31.2, 28.7, 25.1],
                 [17.3, 22.8, 30.5, 26.4, 21.9],
                 [33.1, 29.6, 18.4, 24.0, 27.8],
                 [20.2, 23.5, 31.9, 28.1, 22.6]])
```

**Task 1**:

1. Print the shape of data

(4 weather stations, 5 daily readings each)

```python
# Print shape
print("Shape of data:", data.shape)
```

2.Compute and print mean temperature per station (row-wise mean)

```python
# Mean temperature per station (row-wise mean)
mean_per_station = np.mean(data, axis=1)
print("Mean temperature per station:", mean_per_station)
```

**Task 2**:

Extract all temperature readings above 28.0°C

```python
# Boolean mask for values above 28°C
above_28 = data[data > 28.0]

print("Temperatures above 28°C:", above_28)
```

**Task 3**:

Normalize entire array to range [0, 1]

Formula:
normalized=(data-data.min())/(data.max()-data.min())

```python
# Normalize the data
normalized = (data - data.min()) / (data.max() - data.min())

# Round to 2 decimal places
normalized_rounded = np.round(normalized, 2)

print("Normalized data:\n", normalized_rounded)
```
